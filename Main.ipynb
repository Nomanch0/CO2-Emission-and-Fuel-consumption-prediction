**Scales Numeric columns of Data using StandardScaler**

```python
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import xgboost as xgb

# Assuming you've already scaled and encoded the data, X contains the numeric + encoded columns
X = XGB_data.drop(columns=['Make', 'Model', 'Vehicle Class', 'Transmission', 'Fuel Type'])
y = XGB_data[['Fuel Consumption Comb (L/100 km)','CO2 Emissions(g/km)']] # Target variable
```

**FTTansformer Deep Learning Model**

**Apply PCA to reduce dimensions & Convert To Tensor**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
import torch
from torch.utils.data import Dataset, DataLoader
import torch.nn as nn
import torch.optim as optim
from sklearn.metrics import mean_squared_error, r2_score

# Assuming 'FTTransformer_data' is your dataframe
# Prepare the data
X = FTTransformer_data.drop(columns=['Make', 'Model', 'Vehicle Class', 'Transmission', 'Fuel Type'])
y = FTTransformer_data[['Fuel Consumption Comb (L/100 km)', 'CO2 Emissions(g/km)']]  # Dual target variables

# Apply scaling (StandardScaler for numeric features)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA to reduce dimensionality
pca = PCA(n_components=100)  # Retain 95% of the variance
X_pca = pca.fit_transform(X_scaled)

# Train-test split
X_train, X_val, y_train, y_val = train_test_split(X_pca, y, test_size=0.2, random_state=42)

# Convert data to torch tensors
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.float32)
X_val_tensor = torch.tensor(X_val, dtype=torch.float32)
y_val_tensor = torch.tensor(y_val.values, dtype=torch.float32)
```

**Defines and initializes FTTransformer model**

```python
class FTTransformer(nn.Module):
    def __init__(self, input_dim, output_dim1=1, output_dim2=1):
        super(FTTransformer, self).__init__()
        self.transformer = nn.Transformer(
            d_model=128,
            nhead=4,
            num_encoder_layers=2,
            num_decoder_layers=2,
            dim_feedforward=256,
        )
        self.input_layer = nn.Linear(input_dim, 128)

        # Output layers for both fuel consumption and CO2 emissions
        self.output_layer_1 = nn.Linear(128, output_dim1)  # Fuel Consumption
        self.output_layer_2 = nn.Linear(128, output_dim2)  # CO2 Emissions

    def forward(self, x):
        x = self.input_layer(x)
        x = x.unsqueeze(0)  # Add batch dimension for transformer input
        x = self.transformer(x, x)
        x = x.squeeze(0)  # Remove the batch dimension

        # Get predictions for both outputs
        fuel_consumption_pred = self.output_layer_1(x)
        co2_emissions_pred = self.output_layer_2(x)

        return fuel_consumption_pred, co2_emissions_pred

# Initialize the model
model = FTTransformer(input_dim=X_train.shape[1])
loss_fn = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
# Convert data to DataLoader for batch processing
train_dataset = torch.utils.data.TensorDataset(X_train_tensor, y_train_tensor)
val_dataset = torch.utils.data.TensorDataset(X_val_tensor, y_val_tensor)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32)

# Train the model
num_epochs = 100 # Adjust epochs as necessary

for epoch in range(num_epochs):
    model.train()
    train_loss = 0
    y_train_true_fuel = []
    y_train_pred_fuel = []
    y_train_true_co2 = []
    y_train_pred_co2 = []

    for X_batch, y_batch in train_loader:
        optimizer.zero_grad()

        # Forward pass
        y_pred_fuel, y_pred_co2 = model(X_batch)

        # Compute losses for both predictions
        loss_fuel = loss_fn(y_pred_fuel.squeeze(), y_batch[:, 0])  # Fuel Consumption
        loss_co2 = loss_fn(y_pred_co2.squeeze(), y_batch[:, 1])  # CO2 Emissions

        # Total loss is the sum of the individual losses
        loss = loss_fuel + loss_co2

        loss.backward()
        optimizer.step()

        train_loss += loss.item()

        # Collect true and predicted values for MSE and R2 score
        y_train_true_fuel.extend(y_batch[:, 0].numpy())
        y_train_pred_fuel.extend(y_pred_fuel.squeeze().detach().numpy())

        y_train_true_co2.extend(y_batch[:, 1].numpy())
        y_train_pred_co2.extend(y_pred_co2.squeeze().detach().numpy())

    # Validation step
    model.eval()
    val_loss = 0
    y_val_true_fuel = []
    y_val_pred_fuel = []
    y_val_true_co2 = []
    y_val_pred_co2 = []

    with torch.no_grad():
        for X_batch, y_batch in val_loader:
            y_pred_fuel, y_pred_co2 = model(X_batch)

            loss_fuel = loss_fn(y_pred_fuel.squeeze(), y_batch[:, 0])
            loss_co2 = loss_fn(y_pred_co2.squeeze(), y_batch[:, 1])

            val_loss += loss_fuel.item() + loss_co2.item()

            # Collect true and predicted values for MSE and R2 score
            y_val_true_fuel.extend(y_batch[:, 0].numpy())
            y_val_pred_fuel.extend(y_pred_fuel.squeeze().detach().numpy())

            y_val_true_co2.extend(y_batch[:, 1].numpy())
            y_val_pred_co2.extend(y_pred_co2.squeeze().detach().numpy())

    # Compute RMSE and R2 for both outputs
    train_rmse_fuel = np.sqrt(mean_squared_error(y_train_true_fuel, y_train_pred_fuel))
    train_r2_fuel = r2_score(y_train_true_fuel, y_train_pred_fuel)
    train_rmse_co2 = np.sqrt(mean_squared_error(y_train_true_co2, y_train_pred_co2))
    train_r2_co2 = r2_score(y_train_true_co2, y_train_pred_co2)

    val_rmse_fuel = np.sqrt(mean_squared_error(y_val_true_fuel, y_val_pred_fuel))
    val_r2_fuel = r2_score(y_val_true_fuel, y_val_pred_fuel)
    val_rmse_co2 = np.sqrt(mean_squared_error(y_val_true_co2, y_val_pred_co2))
    val_r2_co2 = r2_score(y_val_true_co2, y_val_pred_co2)

    print(f'Epoch {epoch+1}/{num_epochs} - '
          f'Training RMSE (Fuel): {train_rmse_fuel:.4f}, Training R (Fuel): {train_r2_fuel:.4f}, '
          f'Training RMSE (CO2): {train_rmse_co2:.4f}, Training R (CO2): {train_r2_co2:.4f}, '
          f'Validation RMSE (Fuel): {val_rmse_fuel:.4f}, Validation R (Fuel): {val_r2_fuel:.4f}, '
          f'Validation RMSE (CO2): {val_rmse_co2:.4f}, Validation R (CO2): {val_r2_co2:.4f}')
```

```python
model.eval()
pred_fuel = []
pred_co2 = []
true_fuel = []
true_co2 = []

with torch.no_grad():  # Disable gradient computation for inference
    for X_batch, y_batch in val_loader:
        # Get predictions for both targets
        y_pred_fuel, y_pred_co2 = model(X_batch)

        # Store the predicted values
        pred_fuel.extend(y_pred_fuel.squeeze().cpu().numpy())  # Convert to numpy and store
        pred_co2.extend(y_pred_co2.squeeze().cpu().numpy())

        # Store the true values
        true_fuel.extend(y_batch[:, 0].numpy())  # True fuel consumption
        true_co2.extend(y_batch[:, 1].numpy())

# Calculate RMSE for fuel and CO2
rmse_fuel = np.sqrt(mean_squared_error(true_fuel, pred_fuel))
r2_fuel = r2_score(true_fuel, pred_fuel)

rmse_co2 = np.sqrt(mean_squared_error(true_co2, pred_co2))
r2_co2 = r2_score(true_co2, pred_co2)

# Print RMSE and R for both fuel and CO2
print(f'Validation RMSE (Fuel): {rmse_fuel:.4f}, Validation R (Fuel): {r2_fuel:.4f}')
print(f'Validation RMSE (CO2): {rmse_co2:.4f}, Validation R (CO2): {r2_co2:.4f}')


print(f"Predicted Fuel Consumption : {float(pred_fuel[0]):.2f} L/100km")
print(f"Predicted CO2 Emissions : {float(pred_co2[0]):.2f} g/km")
```

**Stacking Ensembles predictions using meta-model**

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import numpy as np

# Assuming you already have the first model's predictions
y_pred_multi_output = multi_output_model.predict(X_test)

# Second model predictions (your provided code)
pred_fuel = []
pred_co2 = []
true_fuel = []
true_co2 = []

model.eval()
with torch.no_grad():
    for X_batch, y_batch in val_loader:
        # Get predictions for both targets
        y_pred_fuel, y_pred_co2 = model(X_batch)

        # Store the predicted values
        pred_fuel.extend(y_pred_fuel.squeeze().cpu().numpy())  # Convert to numpy and store
        pred_co2.extend(y_pred_co2.squeeze().cpu().numpy())

        # Store the true values
        true_fuel.extend(y_batch[:, 0].numpy())  # True fuel consumption
        true_co2.extend(y_batch[:, 1].numpy())

# Calculate MSE and R for validation
# Calculate RMSE for fuel and CO2
rmse_fuel = np.sqrt(mean_squared_error(true_fuel, pred_fuel))
r2_fuel = r2_score(true_fuel, pred_fuel)

rmse_co2 = np.sqrt(mean_squared_error(true_co2, pred_co2))
r2_co2 = r2_score(true_co2, pred_co2)

# Print RMSE and R for both fuel and CO2
print(f'Validation RMSE (Fuel): {rmse_fuel:.4f}, Validation R (Fuel): {r2_fuel:.4f}')
print(f'Validation RMSE (CO2): {rmse_co2:.4f}, Validation R (CO2): {r2_co2:.4f}')


# Stack predictions from both models
X_stack = np.column_stack([y_pred_multi_output, pred_fuel, pred_co2])

# Assuming y_train_true_fuel and y_train_true_co2 contain the true values
y_train_stack = np.column_stack([true_fuel, true_co2])

# Split the stacked data into training and validation sets
X_train_stack, X_val_stack, y_train_stack, y_val_stack = train_test_split(
    X_stack, y_train_stack, test_size=0.2, random_state=42
)

# Train the meta-model (XGBoost)
model_meta_fuel = xgb.XGBRegressor(objective="reg:squarederror", eval_metric="rmse")
model_meta_co2 = xgb.XGBRegressor(objective="reg:squarederror", eval_metric="rmse")

# Train both meta-models
model_meta_fuel.fit(X_train_stack, y_train_stack[:, 0])
model_meta_co2.fit(X_train_stack, y_train_stack[:, 1])

# Get predictions from the meta-model
y_pred_meta_fuel = model_meta_fuel.predict(X_val_stack)
y_pred_meta_co2 = model_meta_co2.predict(X_val_stack)

# Evaluate the meta-model's performance
val_rmse_meta_fuel = np.sqrt(mean_squared_error(y_val_stack[:, 0], y_pred_meta_fuel))
val_r2_meta_fuel = r2_score(y_val_stack[:, 0], y_pred_meta_fuel)

val_rmse_meta_co2 = np.sqrt(mean_squared_error(y_val_stack[:, 1], y_pred_meta_co2))
val_r2_meta_co2 = r2_score(y_val_stack[:, 1], y_pred_meta_co2)

# Print results
print(f"Meta-model RMSE (Fuel): {val_rmse_meta_fuel:.4f}, Meta-model R (Fuel): {val_r2_meta_fuel:.4f}")
print(f"Meta-model RMSE (CO2): {val_rmse_meta_co2:.4f}, Meta-model R (CO2): {val_r2_meta_co2:.4f}")

# Display some sample predictions
print(f"Predicted Fuel Consumption : {float(y_pred_meta_fuel[0]):.2f} L/100km")
print(f"Predicted CO2 Emissions : {float(y_pred_meta_co2[0]):.2f} g/km")
```
